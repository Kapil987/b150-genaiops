### 📌 Two Important Python Parameters

| Symbol     | Name                           | Meaning                   |
| ---------- | ------------------------------ | ------------------------- |
| `*args`    | Arbitrary positional arguments | many values without names |
| `**kwargs` | Arbitrary keyword arguments    | many values with names    |

They allow a function to accept **unknown number of arguments**. ([w3schools.com][1])

---

## 1️⃣ `*args` (Single `*`)

#### 🧠 Analogy

Think of a **basket of fruits** 🍎🍌🍇

You just **throw fruits in**, no labels.

```text
apple, banana, mango
```

The function collects them in a **tuple**.

### Example

```python
def add_numbers(*numbers):
    total = 0
    for n in numbers:
        total += n
    print("Sum:", total)

add_numbers(2, 4)
add_numbers(1, 2, 3, 4)
```

Output

```
Sum: 6
Sum: 10
```

Meaning

```
numbers = (1,2,3,4)
```

---

# 2️⃣ `**kwargs` (Double `**`)

#### 🧠 Analogy

Think of a **student form** 📋

Everything has a **label**.

```
name = Ram
age = 20
city = Delhi
```

Python collects them in a **dictionary**.

#### Example

```python
def student_info(**data):
    for key, value in data.items():
        print(key, ":", value)

student_info(name="Ram", age=20, city="Delhi")
```

Output

```
name : Ram
age : 20
city : Delhi
```

Meaning

```
data = {
 "name":"Ram",
 "age":20,
 "city":"Delhi"
}
```

---

## 3️⃣ Example Using Both ( Example)

```python
def demo(*args, **kwargs):

    print("Numbers:", args)
    print("Details:", kwargs)

demo(1,2,3, name="Ram", age=25)
```

Output

```
Numbers: (1,2,3)
Details: {'name':'Ram','age':25}
```

---

## 🧠 Simple Rule For Students

```
*args  → many values (tuple)
**kwargs → many named values (dictionary)
```

---

## 🎓 Classroom Trick

Ask students:

```python
demo(1,2,3, name="Ram", age=20)
```

Then ask:

```
Which values go to args?
Which values go to kwargs?
```

Answer

```
args → (1,2,3)
kwargs → {'name':'Ram','age':20}
```
