````python
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)
print("Matrix:")
for row in matrix:
    print(row)
print(f"Element at second row and third column: {matrix[1][2]}")
````

### Assignment 15: Tuple Functions

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

````python
def min_in_tuple(tpl):
    return min(tpl)

def max_in_tuple(tpl):
    return max(tpl)

def sum_of_tuple(tpl):
    return sum(tpl)

sample_tpl = (1, 2, 3, 4, 5)
print(f"Minimum: {min_in_tuple(sample_tpl)}")
print(f"Maximum: {max_in_tuple(sample_tpl)}")
print(f"Sum: {sum_of_tuple(sample_tpl)}")
````

## 📌 Best **Interview Definition** (Remember This)

**Mutable Object**

> An object whose **value can be changed after creation without creating a new object**.

**Immutable Object**

> An object whose **value cannot be changed after creation; a new object must be created instead**. ([Real Python][1])

---

# 🧠 Why Interviewers Use “List inside Tuple” Trick

Because **immutability applies to the container itself, not the objects inside it**.

So:

| Structure | Mutable?    | Why                           |
| --------- | ----------- | ----------------------------- |
| `list`    | ✅ Mutable   | Elements can change           |
| `tuple`   | ❌ Immutable | Elements cannot be reassigned |

But **tuple can hold mutable objects**.

---

# ⚠️ Classic Interview Trick

### Example

```python
t = (1, [2,3])

t[1].append(4)

print(t)
```

Output

```
(1, [2, 3, 4])
```

Interview question:

> “If tuple is immutable, how did the value change?”

### Correct Answer

* Tuple **did not change**
* The **list inside the tuple changed**

Tuple still points to the **same list object**.

---

# 🔬 What Actually Cannot Change

This is what immutability means:

```python
t = (1,2,3)

t[0] = 10
```

❌ Error

Because tuple **cannot reassign its elements**.

---

# 📌 Golden Rule (Best Answer in Interviews)

**Mutable**

```
Object value can change in-place
```

Examples

```
list
dict
set
```

**Immutable**

```
Object value cannot change after creation
```

Examples

```
int
float
str
tuple
```

Lists, dictionaries, and sets are mutable while tuples and strings are immutable in Python. ([GeeksforGeeks][2])

---

# 🧠 One-Line Interview Answer

> **Mutability depends on whether the object itself can change in memory, not whether objects inside it are mutable.**

---

# 💡 Best 5-Second Example (Use in Interview)

```python
a = [1,2,3]
a[0] = 10   # allowed → mutable

b = (1,2,3)
b[0] = 10   # error → immutable
```
