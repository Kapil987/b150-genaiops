```python
## List comprehension, 2^2 = 2*2 = 4, 2^3 = 2*2*2= 8, 3^2 = 3*3= 9
# iteration 1 , range = 0 , x = 0 --> 0^2 =0 , lst[0]
# iteration 2 , range = 1 , x = 1 --> 1^2 =1 , lst[0,1]
# iteration 3 , range = 2 , x = 2 --> 2^2 =4 , lst[0,1,4]
# iteration 4 , range = 3 , x = 3 --> 3^2 =9 , lst[0,1,4,9]
# iteration 5 , range = 4 , x = 4 -- 4^2 =16 , lst[0,1,4,9,16]

lst=[]
for x in range(10):
    # print("inside for loop: ",lst)
    lst.append(x**2) # 4^2 = 4 * 4 = 16

print("outside for loop",lst)
```

##### List Comprehension

Basics Syantax            [expression for item in iterable]

with conditional logic    [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]




```python
### Basic List Comphrension

sqaure=[num**2 for num in range(10)]
print(sqaure)
```

```python
### List Comprehension with Condition
# iteration 1 , range = 0 , i = 0 
# iteration 2 , range = 1 , i = 1
# iteration 3 , range = 2 , i = 2
# iteration 4 , range = 3 , i = 3
# iteration 5 , range = 4 , i = 4

lst=[]
for i in range(10):
    if i%2==0:              # i = 2 true
        lst.append(i)       # i = 2

print(lst)
        
```

```python
even_numbers= [ num for num in range(10) if num%2==0 ]

print(even_numbers)
```

```python
## List Comprehension with function calls , [5, 5, 6, 4, 13]
words = ["hello", "world", "python", "list", "comprehension"]

# print(len("python"))

for word in words:
  # print(word,len(word))
  print(len(word),end=' ')

# lengths = [ len(word) for word in words  ]
# print(lengths)  # Output: [5, 5, 6, 4, 13]

```
