```python
class Car:
  def __init__(self,windows,doors,enginetype):
    self.windows= windows
    self.doors= doors
    self.enginetype= enginetype
    print('Parent class constructor was called')
  def drive(self):
    print(f'the person will drive the {self.enginetype} car')

car1 = Car(4,5,'petrol')
car1.drive()
```

```python
class Tesla(Car):
  def __init__(self, windows, doors, enginetype,is_selfdriving):
    super().__init__(windows, doors, enginetype)
    self.is_selfdriving=is_selfdriving
  
  def selfdriving(self):
    print(f'Tesla support self driving: {self.is_selfdriving}')

tesla1=Tesla(6,4,'electric',True)

tesla1.selfdriving()
```

```python
class Animal:
  def speak(self):
    return 'sound of the animal'

# derived class
class Dog(Animal):
  def speak(self):
    return "Woof!"
    # return super().speak()

class Cat(Animal):
  def speak(self):
    return 'Meow!'

def animal_speak(input_animal):
  print('The input animal says -->',input_animal.speak())

dog = Dog()
cat = Cat()
dog.speak()

animal_speak(cat)
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
  def drive(self):
    print('the vehicle is used for driving')

  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  def start_engine(self):
    print('car engine started')
  
  def demo_function(self):
    print('this is a demo function')

def operate_vehicle(vehicle):
  vehicle.drive()

car=Car()

operate_vehicle(car)
```
