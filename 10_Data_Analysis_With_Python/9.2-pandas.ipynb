#### Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. 
- A Series is a one-dimensional array-like object, 
- while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

**Difference between `df.loc` and `df.iloc`**

---

### 1. Selection basis
- **`df.loc`**  
  - Selects by **label** (row and column names).  
  - Includes both start and end of a slice.  
- **`df.iloc`**  
  - Selects by **integer position** (zero-based).  
  - End of a slice is **exclusive**.
---

### 3. Quick takeaways
- Use **`.loc`** when you know the **index/column names**.  
- Use **`.iloc`** when you want purely **positional** selection.

```python
# 1.Sample DataFrame
df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60],
    'C': [70, 80, 90]
}, index=['row1', 'row2', 'row3'])

print(df)
```

```python
#### a) Single-element access
#       0    1   2
#        A   B   C
# row1  10  40  70   # 0
# row2  20  50  80   # 1
# row3  30  60  90   # 2

# - **By label**:  
# df.loc['row2', 'B']      # → 50

# - **By position**:  
# df.iloc[1, 1]            # → 50 , direct print outputs in numpy format, print produces usable outputs
print(df.iloc[1,1])
# print(type(df.iloc[1,1]))
```

```python
#### b) Row slicing
# - **Labels** (inclusive):  
# df.loc['row1':'row2']

# - **Positions** (end exclusive):  
df.iloc[0:2]    # start : end-1 
```

```python
#### c) Column slicing
# - **Labels** (inclusive):  
# df.loc[:, 'A':'B']  #   : no restriction

# - **Positions** (end exclusive):  
df.iloc[:, 0:2] # start : end-1
```

In Pandas, these two parameters are essential for controlling **how** and **where** your data is modified.

### 1. `axis=1` (The Direction)
The `axis` parameter tells Pandas whether you want to drop a **row** or a **column**.

* **`axis=0` (Default):** Refers to the **horizontal axis (rows)**. If you use this, Pandas looks for index labels to remove.
* **`axis=1`:** Refers to the **vertical axis (columns)**. By setting `axis=1`, you are telling Pandas: "Look at the column headers and drop the one named 'Salary'."

> **Analogy:** Think of `axis=0` as a row of houses on a street, and `axis=1` as the vertical pillars holding up a roof.



---

### 2. `inplace=True` (The Storage)
By default, most Pandas operations **return a new object** and leave the original DataFrame untouched. `inplace` decides if the change happens "on the spot."

* **`inplace=False` (Default):** * The original `df` remains unchanged.
    * It creates a "copy" with the column removed.
    * *Usage:* `new_df = df.drop('Salary', axis=1)`
* **`inplace=True`:** * The operation modifies the original `df` directly.
    * It returns `None`. 
    * *Usage:* `df.drop('Salary', axis=1, inplace=True)`

---

### Summary Table

| Parameter | Value | What it does |
| :--- | :--- | :--- |
| **axis** | `1` | Targets **Columns** instead of Rows. |
| **inplace** | `True` | Updates the **original variable** directly; no need to re-assign. |

**Pro-Tip:** Modern Pandas practice is actually moving away from `inplace=True`. Many developers now prefer re-assigning the variable (e.g., `df = df.drop(...)`) because it is clearer and avoids certain "SettingWithCopy" warnings.


```python
## Remove a column
df.drop('Salary',axis=1)
# df.drop('Salary',axis=1,inplace=True)
```
