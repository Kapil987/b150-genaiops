```python
## Arithmethic Operation
# 2x2x2
a= 2
b= 3
# a= 16
# b= 4

add_result=a+b  #addiiton
sub_result=a-b  #substraction 
mult_result=a*b #multiplication
div_result=a/b  #division
floor_div_result=a//b ## floor division --> quotient
modulus_result=a%b #modulus operation -- remainder

exponent_result=a**b ## Exponentiation


print(add_result)
print(sub_result)
print("mult_result: ",mult_result)
print("div_result: ",div_result)
print("floor_div_result: ",floor_div_result)
print("modulus_result: ",modulus_result)
print("exponent_result: ",exponent_result)

# Dividend = (Divisor \times Quotient) + Remainder
```

#### Logical Operators

```python
## And ,Not,OR
X=True # 2+3*4, 1
Y=True # 4-7+6 , 1

result =X and Y
print(result)
```

```python
X=False
Y=True

result =X and Y
print(result)
```

```python
## OR
X=True # 1
Y=False # 0

result =X or Y
print(result)
```

```python
# Not operator
X=False
not X
```

```python
## Simple Calculator

# Simple calculator
num1 = float(input("Enter first number: ")) # float (abc) --> invalid operation
num2 = float(input("Enter second number: "))

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2

# Displaying results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```
