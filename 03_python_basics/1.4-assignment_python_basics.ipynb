Python is mainly an interpreted language, but internally it is both compiled and interpreted.

### How Python Actually Runs

1. Python source code (`.py`) is first compiled into bytecode (`.pyc`). 
2. The Python Virtual Machine (PVM) then interprets that bytecode and executes it. 

### One-Liner

“Python is considered an interpreted language, but internally it first compiles code to bytecode and then interprets it using the Python Virtual Machine.”

### Quick Flow

```
Python code (.py)
        ↓
Compile to Bytecode (.pyc)
        ↓
Python Virtual Machine
        ↓
Program Output
```
