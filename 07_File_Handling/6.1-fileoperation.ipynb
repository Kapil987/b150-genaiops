#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
### Read a Whole File

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## sstrip() removes the newline character -> \n
        # print(line) ## sstrip() removes the newline character
```

```python
## Writing a file(Overwriting)

with open('example.txt','w') as file:
    file.write('Hello GenAI!\n')
    file.write('this is a new line.')
```

```python
## Write a file(without Overwriting)
with open('example.txt','a') as file:
    file.write("Append operation taking place!\n")
```

```python
### Writing a list of lines to a file
lines=['First line \n','Second line \n','Third line\n']

with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from a source text file and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()
    print(content)

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
#Read a text file and 
# 1) count the number of lines, 
# 2) words, and 
# 3) characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path): # file_path = example.txt
    with open(file_path, 'r') as file:
        lines = file.readlines()
        print('lines -->',lines)
        line_count = len(lines)

        # print('lines count',line_count)
        # mycount=0
        # for line in lines: # iter 1 line =i am using windows laptop , mycount =12 
        #     print('line.split',line.split()) # 'i', 'am', 'using', 'windows', 'laptop'
        #     print('len',len(line.split())) # 5
        #     mycount = mycount + len(line.split())
        #             # 7 + 5
        # print('sum of len count',mycount)

    #     # short line code below
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count
    # return line_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
# lines = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
# print(f'Lines: {lines}')
```

The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten).

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world11\n")
    file.write("This is a new line 11\n")

    ## Move the file cursor to the beginning
    file.seek(5)
    
    ## Read the content of the file
    content=file.read()
    print(content)
```
