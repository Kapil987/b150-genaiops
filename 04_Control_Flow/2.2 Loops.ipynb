#### Analogy
A for loop is like checking items on a shopping list:

- You have a specific list of items to buy
- You look at each item one by one
- You cross off each item as you go
- You finish when you've gone through every item on the list

A while loop is like washing dishes:

- You check if there are dirty dishes in the sink
- You keep washing as long as dirty dishes remain
- You stop when there are no more dirty dishes
- You don't know at the start exactly how many dishes you'll wash

````python
## while loop

## The while loop continues to execute as long as the condition is True.

count=0

while count<3: #
    print(count)
    count=count+1
print("count finished")

# iteration
# count 0 --> 0<3: --> True, count=0+1, final count value is --> 1
# count 1 --> 1<3: --> True, count=1+1, final count value is --> 2
# count 2 --> 2<3: --> True, count=1+1, final count value is --> 3
# count 3 --> 3<3: --> False, count=1+1, final count value is --> 


````

````python
## Loop Control Statements

## break
## The break statement exits the loop permaturely

## break sstatement

for i in range(10): # start from 0, max would be 10-1 = 9
    if i==4: # 
        break
    print(i)
print("loop finished")
# i = 0 --> if 0==4 -- false
# i = 4 --> if 4==4 -- True
````

````python
## continue

## The continue statement skips the current iteration and continues with the next.

for i in range(10): # i 0, i max value 9  
    if i%2==0:      # 0/2 --> 0 == 0
        continue
    print(i)

# 0/2 --> 0 == 0 , True , execute continue statement
# 1/2 --> 1/2 == 0 , false , execute continue statement
# 2/2 --> 2/2 == 0 , True , execute continue statement
# 3/2 --> 3/2 == 0 , False , execute continue statement

````

````python
## pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i==3: # false
        pass
    print(i)
````

````python
## Nested loopss
## a loop inside a loop
# i start value is 0, max value is 2
# ij start value is 0, max value is 1

for i in range(3):              # i = 0, i = 0, i = 1, i = 1, i =2, i = 2 , 
    for j in range(2):          # j = 0, j = 1, j = 0 , j = 1, j =0 , j =1, 
        print(f"i:{i} and j:{j}")

# itr 1 --> i = 0
````

````python
## Examples- Calculate the sum of first N natural numbers using a while and for loop

## while loop  

n=10   
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1

print("Sum of first 10 natural number:",sum)
````

````python
## Example- Prime numbers between 1 and 100

for num in range(1,101): 
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
````

## 📘 Concept: Logical Operators

Python has **three logical operators** used to combine conditions:

| Operator | Meaning                                    |
| -------- | ------------------------------------------ |
| `and`    | True **only if both conditions are true**  |
| `or`     | True if **at least one condition is true** |
| `not`    | Reverses the result                        |

Logical operators return **True or False** and are commonly used in `if` statements to control program flow. ([W3Schools][1])
